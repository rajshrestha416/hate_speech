```python
#imports libraries
import numpy as np
import pandas as pd
```

```python
#import data
df_twitter = pd.read_csv("twitter_hate_speech.csv")
df_nepali = pd.read_excel("nepali_hate_speech.xlsx")
```

```python
df_twitter.info()
```

```python
df_nepali.info()
```

```python
def map_class_by_offensiveness(offensiveness):
    if offensiveness in (1, 2):
        return 2
    elif offensiveness in (3, 4):
        return 0
    else:
        return 1
```

```python
df_nepali['Offensiveness'] = df_nepali['Offensiveness'].apply(map_class_by_offensiveness)
```

```python
df_nepali
```

```python
# Nepali
df_nepali_preprocess = df_nepali.rename(columns={'RawRom': 'text', 'Offensiveness': 'label'})
df_nepali_preprocess['category'] = df_nepali['Offensiveness'].map({0: 'hate_speech', 1: 'offensive_language', 2: 'neither'})
```

```python
df_nepali_preprocess
```

```python
data_nepali = df_nepali_preprocess[['text', 'label', 'category']]
```

```python
data_nepali
```

```python
df_twitter
```

```python
# twitter
df_twitter_preprocess = df_twitter.rename(columns={'tweet': 'text', 'class': 'label'})
df_twitter_preprocess['category'] = df_twitter['class'].map({0: 'hate_speech', 1: 'offensive_language', 2: 'neither'})
```

```python
df_twitter_preprocess
```

```python
data_twitter = df_twitter_preprocess[['text', 'label', 'category']]
```

```python
data_twitter
```

```python
data = pd.concat([data_twitter, data_nepali], ignore_index=True)
```

```python
data
```

```python
data.to_csv('merge_data.csv', index=False, encoding='utf-8')
```
